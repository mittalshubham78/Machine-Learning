```python
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random


data = load_iris()
X = data['data']
y = data['target']

X
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# Number of training data
n = X.shape[0]
# Number of features in the data
c = X.shape[1]

# Plot of the dataset
%matplotlib notebook
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimension
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
X_reduced = PCA(n_components=3).fit_transform(X)
# print(X_reduced)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],c=y, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
import numpy as np
from copy import deepcopy
# Number of clusters
#
wcss_list = []
for K in range(1,11):    
    centers = np.array([]).reshape(c,0)

    for i in range(K):
        rand = random.randint(0,n-1)
        print(rand)
        centers = np.c_[centers,X[rand]]

    centers  = centers.transpose()
    centers = np.array(centers)

    centers_old = np.zeros(centers.shape) # to store old centers
    centers_new = deepcopy(centers) # Store new centers

    clusters = np.zeros(n)
    distances = np.zeros((n,K))
    centers_new
    
    error = np.linalg.norm(centers_new - centers_old)

# When, after an update, the estimate of that center stays the same, exit loop
    while error != 0:
    # Measure the distance to every center
        for i in range(K):
            distances[:,i] = np.linalg.norm(X - centers_new[i], axis=1)
    # Assign all training data to closest center
        clusters = np.argmin(distances, axis = 1)
    
        centers_old =deepcopy(centers_new)
    # Calculate mean for every cluster and update the center
        for i in range(K):
            centers_new[i] = np.mean(X[clusters == i], axis=0)
        error = np.linalg.norm(centers_new - centers_old)
        
    min_dist_list=[]
    def wcss(distances):
        min_dist_list.append((distances.min(axis=1))**2)
        wcss_value=np.sum(min_dist_list)
        return wcss_value

       
    wcss_value=wcss(distances)
    wcss_list.append(wcss_value)
    print(wcss_value)

#print(wcss_list)
arr=np.array(wcss_list)
print(arr)
```
